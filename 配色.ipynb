```
install.packages("ggplot2")

library(ggplot2)
```

```
NBA1516<-fetch_NBAPlayerStatistics("15-16")
qplot(TotalMinutesPlayed,TotalPoints,data=NBA1516,geom=c("point","smooth"))

#總出場時間跟總得分的點與漸進線圖
```

```
qplot(TotalMinutesPlayed,data=NBA1516,fill=Position)
#總出場時間的堆疊長條圖
```

```
qplot(TotalMinutesPlayed,TotalPoints,data=NBA1516,facets=.~Position)

#總出場時間跟總得分的散布圖
```

```
ggplot(data=NBA1516,aes(y=TotalPoints,x=TotalMinutesPlayed))+geom_violin()

#總出場時間跟總得分的小提琴圖
```
